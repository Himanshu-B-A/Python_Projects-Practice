```python
import cv2
import matplotlib.pyplot as plt

# Read the image in color mode
image_path = r'All_Image\test_cv.jpeg'
img_retest = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Check if image is loaded
if img_retest is None:
    print("Error: Image not found.")
else:
    # Split the channels
    b, g, r = cv2.split(img_retest)

    # Set up figure size
    plt.figure(figsize=[20, 5])

    # Correcting subplot function usage
    plt.subplot(141); plt.imshow(r, cmap='Reds'); plt.title("Red Channel")
    plt.subplot(142); plt.imshow(g, cmap='Greens'); plt.title("Green Channel")
    plt.subplot(143); plt.imshow(b, cmap='Blues'); plt.title("Blue Channel")

    # Merge back the channels
    imgMerge = cv2.merge((b, g, r))

    # Display the merged image
    plt.subplot(144)
    plt.imshow(imgMerge[:, :, ::-1])  # Convert BGR to RGB for correct display
    plt.title("Merged Image")

    # Show the plots
    plt.show()
```
